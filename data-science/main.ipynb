```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;

traffic_data = pd.read_csv('./dataset_flujo_vehicular.csv');
filter_columns = traffic_data[['CANTIDAD', 'LATITUD', 'LONGITUD']];

print(traffic_data.describe());

#Comprobar que no haya datos vacios en mi dataset
print(filter_columns.info());
filter_columns.describe();
print(filter_columns.isnull().count())


cantidad_total = filter_columns.groupby(['LATITUD', 'LONGITUD']).agg('sum');

cantidad_total.to_csv("traffic_data.csv");
Y_labels = [];

for i in range(len(cantidad_total.index)):
    Y_labels.append(i);


fig, (ax1, ax2) = plt.subplots(1,2);

ax1.bar(Y_labels, cantidad_total['CANTIDAD'])

ax2.boxplot(filter_columns['CANTIDAD']);

plt.show();

```
